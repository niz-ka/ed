## Spam detection using Naive Bayes

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
```

```python
df = pd.read_csv('smsspamcollection.csv', sep='\t', header=None, names=['label', 'text'])
print(df.shape)
df.head(10)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df.text)
tokens_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(tokens_df.shape)
tokens_df.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(tokens_df, df.label, test_size=0.3, random_state=0)
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
